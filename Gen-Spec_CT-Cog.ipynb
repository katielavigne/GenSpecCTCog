### Gen-Spec CT-Cog Associations w/ CPZ adherence, functioning, affective diagnosis, & substance dependence

```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = 'Times New Roman'
plt.rcParams.update({'font.size': 12})
```

```python
datadir = 'C:/Users/lavig/OneDrive - McGill University/research/publishing/inprep_Lavigne_EurNPP_CTCognition/data/'
df1 = pd.read_csv('data1_glimfile_pls.csv')
df2 = pd.read_csv('data2_glimfile_pls.csv')
```

## 1. Global CT & Cognition LV1

### CPZ_Adherence

```python
#sample 1
x=df1.gen_gen1
y=df1.cpz_adherence
res1 = stats.pearsonr(x,y)
sns.regplot(x=x,y=y)

#sample 2
df2_cpz = df2.dropna(subset=['cpz_adherence'])
x=df2_cpz.gen_gen1
y=df2_cpz.cpz_adherence
res2 = stats.pearsonr(x,y)

print('Global CT LV1 - Gen Cog\n S1: r(' 
      + str(len(df1)-2) 
      + ') = ' 
      + str(round(res1[0],2))
      + ', p = '
      + str(round(res1[1],2))
      + '\n S2: r('
      + str(len(df2_cpz)-2) 
      + ') = ' 
      + str(round(res2[0],2))
      + ', p = '
      + str(round(res2[1],2)))

ax = sns.regplot(x=x,y=y)
ax.set(xlabel='Global Thickness LV1 - General Cognition', ylabel='CPZ Adherence')
plt.legend(labels=['Sample 1','_','_','Sample 2','_','_'])
```

### Duration of Illness

```python
#sample 1
x=df1.gen_gen1
y=df1.dur_illness
res1 = stats.pearsonr(x,y)
sns.regplot(x=x, y=y)

#sample 2
df2_dur_ill = df2.dropna(subset=['dur_illness'])
df2_dur_ill = df2_dur_ill[df2_dur_ill.dur_illness < 8] #remove outliers
x=df2_dur_ill.gen_gen1
y=df2_dur_ill.dur_illness
res2 = stats.pearsonr(x,y)

print('Global CT LV1 - Gen Cog\n S1: r(' 
      + str(len(df1)-2) 
      + ') = ' 
      + str(round(res1[0],2))
      + ', p = '
      + str(round(res1[1],2))
      + '\n S2: r('
      + str(len(df2_dur_ill)-2) 
      + ') = ' 
      + str(round(res2[0],2))
      + ', p = '
      + str(round(res2[1],2)))

ax = sns.regplot(x=x,y=y)
ax.set(xlabel='Global Thickness LV1 - General Cognition', ylabel='Duration of Illness')
plt.legend(labels=['Sample 1','_','_','Sample 2','_','_'])
```

### Functioning

```python
#sample 1
df1_func = df1.dropna(subset=['sofas'])
x=df1_func.gen_gen1
y=df1_func.sofas
res1 = stats.pearsonr(x,y)
sns.regplot(x=x, y=y)

#sample 2
df2_func = df2.dropna(subset=['sofas'])
x=df2_func.gen_gen1
y=df2_func.sofas
res2 = stats.pearsonr(x,y)

print('Global CT LV1 - Gen Cog\n S1: r(' 
      + str(len(df1_func)-2) 
      + ') = ' 
      + str(round(res1[0],2))
      + ', p = '
      + str(round(res1[1],2))
      + '\n S2: r('
      + str(len(df2_func)-2) 
      + ') = ' 
      + str(round(res2[0],2))
      + ', p = '
      + str(round(res2[1],2)))

ax = sns.regplot(x=x,y=y)
ax.set(xlabel='Global Thickness LV1 - General Cognition', ylabel='Functioning')
plt.legend(labels=['Sample 1','_','_','Sample 2','_','_'])
```

### Affective diagnosis

```python
# sample 1
t1=df1.gen_gen1[df1.dx_aff==0]
t2=df1.gen_gen1[df1.dx_aff==1]

print(stats.levene(t1,t2))
print(stats.ttest_ind(t1,t2,equal_var=True))

ax = sns.violinplot(x=df1.dx_aff,y=df1.gen_gen1)
ax.set(xlabel='Sample 1 Global Thickness LV1 - General Cognition', ylabel='Affective/Non-Affective Diagnosis')
plt.legend(labels=['Non-Affective','_','_','_','Affective'], loc=9)
```

```python
# sample 2
df2_aff = df2.dropna(subset=['dx_aff'])
t1=df2_aff.gen_gen1[df2_aff.dx_aff==0]
t2=df2_aff.gen_gen1[df2_aff.dx_aff==1]

print(stats.levene(t1,t2))
print(stats.ttest_ind(t1,t2,equal_var=True))

ax = sns.violinplot(x=df2_aff.dx_aff,y=df2_aff.gen_gen1, labels=[''])
ax.set(xlabel='Sample 2 Global Thickness LV1 - General Cognition', ylabel='Affective/Non-Affective Diagnosis')
plt.legend(labels=['Non-Affective','_','_','_','Affective'], loc=9)
```
