# Lavigne et al. (2022) General and Specific Cortical Thickness-Cognition Associations in First-Episode Psychosis
## Associations between PLS brain scores and clinical measures

```python
# import libraries
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# set defaults
warnings.filterwarnings('ignore')
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = 'Times New Roman'
plt.rcParams.update({'font.size': 12})
```

```python
# data prep
datadir = 'C:/Users/lavig/OneDrive - McGill University/research/publishing/inprep_Lavigne_EurNPP_CTCognition/data/'
df1 = pd.read_csv(datadir + 'data1_glimfile_pls.csv')
df2 = pd.read_csv(datadir + 'data2_glimfile_pls.csv')
```

## 1. Global CT & Cognition LV1

### Chlorpromazine equivalent weighted by adherence

```python
# sample 1
x = df1.genCTcog1
y = df1.cpz_adherence
res1 = stats.pearsonr(x,y)
sns.regplot(x=x,y=y)

# sample 2
df2_cpz = df2.dropna(subset=['cpz_adherence'])
x = df2_cpz.genCTcog1
y = df2_cpz.cpz_adherence
res2 = stats.pearsonr(x,y)

# results
print('Global CT LV1 (Gen Cog) w/ CPZ Adherence\n S1: r(' 
      + str(len(df1)-2) 
      + ') = ' 
      + str(round(res1[0],2))
      + ', p = '
      + str(round(res1[1],2))
      + '\n S2: r('
      + str(len(df2_cpz)-2) 
      + ') = ' 
      + str(round(res2[0],2))
      + ', p = '
      + str(round(res2[1],2)))

# figure
ax = sns.regplot(x=x,y=y)
ax.set(ylabel='Chlorpromazine Equivalent Weighted by Adherence', xlabel='Latent Variable Brain Score', title='GLOBAL THICKNESS LV1 - GENERAL COGNITION')
plt.legend(labels=['_','_','Sample 1','_','Sample 2','_','_'])
```

### Duration of Illness

```python
#sample 1
x = df1.genCTcog1
y = df1.dur_illness
res1 = stats.pearsonr(x,y)
sns.regplot(x=x, y=y)

#sample 2
df2_dur_ill = df2.dropna(subset=['dur_illness'])
#df2_dur_ill = df2_dur_ill[df2_dur_ill.dur_illness < 8] # remove outliers
x = df2_dur_ill.genCTcog1
y = df2_dur_ill.dur_illness
res2 = stats.pearsonr(x,y)

# results
print('Global CT LV1 (Gen Cog) w/ Duration of Illness\n S1: r(' 
      + str(len(df1)-2) 
      + ') = ' 
      + str(round(res1[0],2))
      + ', p = '
      + str(round(res1[1],2))
      + '\n S2: r('
      + str(len(df2_dur_ill)-2) 
      + ') = ' 
      + str(round(res2[0],2))
      + ', p = '
      + str(round(res2[1],2)))

# figure
ax = sns.regplot(x=x,y=y)
ax.set(ylabel='Duration of Illness', xlabel='Latent Variable Brain Score', title='GLOBAL THICKNESS LV 1 - GENERAL COGNITION')
plt.legend(labels=['_','_','Sample 1','_','Sample 2','_','_'])
```

### Functioning

```python
# sample 1
df1_func = df1.dropna(subset=['sofas'])
x = df1_func.genCTcog1
y = df1_func.sofas
res1 = stats.pearsonr(x,y)
sns.regplot(x=x, y=y)

# sample 2
df2_func = df2.dropna(subset=['sofas'])
x = df2_func.genCTcog1
y = df2_func.sofas
res2 = stats.pearsonr(x,y)

print('Global CT LV1 (Gen Cog) w/ Functioning\n S1: r(' 
      + str(len(df1_func)-2) 
      + ') = ' 
      + str(round(res1[0],2))
      + ', p = '
      + str(round(res1[1],2))
      + '\n S2: r('
      + str(len(df2_func)-2) 
      + ') = ' 
      + str(round(res2[0],2))
      + ', p = '
      + str(round(res2[1],2)))

ax = sns.regplot(x=x,y=y)
ax.set(xlabel='Functioning', ylabel='Latent Variable Brain Score', title='GLOBAL THICKNESS LV 1 - GENERAL COGNITION')
plt.legend(labels=['_','_','Sample 1','_','Sample 2','_','_'])
```
